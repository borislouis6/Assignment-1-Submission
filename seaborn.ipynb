Q. Implement Seaborn Plots on Iris dataset

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
%matplotlib inline
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
iris.head()
```

```python
iris["species"].value_counts()
```

```python
iris.plot(kind='scatter', x="sepal_length", y="sepal_width")
```

```python
iris.plot(kind='scatter', x="petal_length", y="petal_width")
```

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, size=6)
```

```python
sns.jointplot(x='petal_length', y='petal_width', data=iris, height=6)
```

```python
sns.FacetGrid(iris, hue="species", palette='husl', height=6)\
     .map(plt.scatter, "sepal_length", "sepal_width")\
          .add_legend()
```

```python
sns.FacetGrid(iris, hue="species", palette='husl', height=6)\
     .map(plt.scatter, "petal_length", "petal_width")\
          .add_legend()
```

```python
sns.pairplot(iris, hue='species', height=4)
```
